```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train
```

```python
X_train.shape   # 60k images of 28 X 28 resolution
```

```python
X_train[0]
```

```python
X_test.shape
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_prob
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
model_2 = Sequential()
model_2.add(Flatten(input_shape=(28,28)))
model_2.add(Dense(128,activation='relu'))
model_2.add(Dense(32,activation='relu'))
model_2.add(Dense(10,activation='softmax'))
```

```python
model_2.summary()
```

```python
model_2.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model_2.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob_2 = model_2.predict(X_test)
```

```python
y_pred_2 = y_prob_2.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
